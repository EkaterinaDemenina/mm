$x'=k(x)*x$

мягкая модель Мальтуса с зависящим от населения коэффициентом размножения

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
        a=0.6
        b=0.1
        return a*y-b*y*y
```

```python
t=np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = ay-by^2$")
plt.grid(True)
plt.legend()
plt.show()
```

$y'=y-y^2-c$

жесткая модель рыболовства с интенсивностью с (с=1/4 - критическое значение), где х - количество рыб в озере или мировом океане

```python
def dydt(y,t):
    c = 0.42
    return y-y*y-c
```

```python
t=np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = y-y^2-c$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=x-g(x)x-c$

мягкая модель рыболовства, где критическое значение с является максимумом функции g(x)x

```python
def dydt(y,t):
    a=0.6
    b=0.3
    c=0.2
    return y-(a-b*y)*y-c
```

```python
t=np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = y-(a-b*y)*y-c$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=x-x^2-kx$

модель изменения популяции рыб при дифференциальной квоте вылова

```python
def dydt(y,t):
    k=0.6
    return y-y*y-k*y
```

```python
t=np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = y-y*y-k*y$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=kx$

жесткая модель роста населения Земли (модель Мальтуса)

```python
def dydt(y, t):
        k = 0.42
        return -k * y
```

```python
t = np.linspace(0, 10, 100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = -ky$")
plt.grid(True)
plt.legend()
plt.show()
```
